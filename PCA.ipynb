```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
# Generate a random dataset of 200 samples with 10 features
X, _ = make_blobs(n_samples=200, centers=5, n_features=10, random_state=42)

# Apply PCA to reduce the dimensionality of the dataset
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the original dataset in 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], s=40)
ax.set_title('Original Dataset in 3D')
plt.show()

# Plot the original dataset in 2D
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], s=40)
ax.set_title('Original Dataset in 2D')
plt.show()

# Plot the PCA-reduced dataset in 2D
fig, ax = plt.subplots()
ax.scatter(X_pca[:, 0], X_pca[:, 1], s=40)
ax.set_title('PCA-Reduced Dataset in 2D')
plt.show()

# Print the explained variance ratio of each principal component
print('Explained Variance Ratio:', pca.explained_variance_ratio_)
```

In this example, we first generated a random dataset of 200 samples with 10 features using the `make_blobs` function from `sklearn.datasets`. Then, we applied PCA to the dataset to reduce the dimensionality from 10 to 2 features using `sklearn.decomposition.PCA`.

We then plotted the original dataset in 3D and 2D using `matplotlib`. We also plotted the PCA-reduced dataset in 2D. Finally, we printed the explained variance ratio of each principal component, which indicates the amount of variance in the original dataset that is explained by each principal component.

The first plot shows the original dataset in 3D, with each point representing a sample and the color indicating the cluster. The second plot shows the same dataset in 2D, with only the first two features. The third plot shows the PCA-reduced dataset in 2D, with the two principal components.

The explained variance ratio shows that the first principal component explains 62.8% of the variance in the dataset, while the second principal component explains 22.6% of the variance. This means that the two principal components together explain 85.4% of the variance in the dataset. 

Overall, PCA was able to successfully reduce the dimensionality of the dataset while retaining most of the important information.
